```python
#Create independent and Dependent Features
columns = df_resampling.columns.tolist()
# Filter the columns to remove data we do not want 
columns = [c for c in columns if c not in ['Churn']]
# Store the variable we are predicting 
target = 'Churn'
# Define a random state 
state = np.random.RandomState(42)
X = df_resampling[columns]
Y = df_resampling[target]
# Print the shapes of X & Y
print(X.shape)
print(Y.shape)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter = 359, random_state = 42)
model_lr = log_reg.fit(x_train, y_train)
y_pred_lr = model_lr.predict(x_test)

print('confusion matrix:','\n', confusion_matrix(y_test, y_pred_lr))
print('accuracy score:',accuracy_score(y_test, y_pred_lr))
print('precision score:',precision_score(y_test, y_pred_lr))
print('recall score:',recall_score(y_test, y_pred_lr))
```

    -> KNN MODEL.

```python
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings("ignore")

error1= []
error2= []
for k in range(1,20):
    knn= KNeighborsClassifier(n_neighbors=k)
    model1_knn = knn.fit(x_train,y_train)
    y_pred_knn_1= model1_knn.predict(x_train)
    error1.append(np.mean(y_train!= y_pred_knn_1))
    y_pred_knn_2= model1_knn.predict(x_test)
    error2.append(np.mean(y_test!= y_pred_knn_2))


plt.plot(range(1,20), error1, 'o-', label = 'Train Data')
plt.plot(range(1,20), error2, 'o-', label = 'Test Data')
plt.xlabel('k-Value')
plt.ylabel('Error')
plt.legend()
plt.grid()
```

```python
k_matrix = []
neighbors = np.arange(1, 20)

for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    model2_knn = knn.fit(x_train, y_train)
    y_pred2_knn = model2_knn.predict(x_test)
    acc = accuracy_score(y_test,y_pred2_knn)
    k_matrix.append(acc)

plt.plot(neighbors, k_matrix,'o-')
plt.xlabel('k-Value')
plt.ylabel('Accuracy')
plt.grid()
```

NOTE:- From the above two plots(Error and Accuracy) it could be concluded that the optimum value of k lies below 2.5, so "1" is taken as the final value for model deployement.
